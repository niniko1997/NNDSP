# Overview

The question we are trying to answer is the following:  

1. Is distribution of ages the main driver for performance differences?

To successfully answer this question, we have devised some methods to synthetically recreate train and test set age distributions.
- Draw subjects with same age distribution as NNDSP from NKI data
- Use training dataset with the same distribution as HCP data
- Training set as uniform as possible in age and test how that performs
- Take HCP age range distribution from NKI to test if performance is just as bad

# Age Distributions

```python
def get_bucket_size(arr):
    import scipy
    
    iqr = scipy.stats.iqr(arr)
    return 2*iqr / (len(arr)**(1/3))
```

```python
def num_subj_in_bucket(df, nosub, index, bucket, b_size=5):
    # calculates the number of subjects of bucket size b_size in bucket
    noages = 0
    while index < nosub and float(df.age.values[index]) <= bucket + bucket_size:
        noages +=1
        index +=1
    return noages, index
```

```python
def find_dist(df, minimum = 0, maximum = 100, bucket_size = 5):
    index = 0
    age_dist = {}
    
    for x in np.arange(minimum, maximum, bucket_size):
        # count how many subjects are in the age bucket (given df is sorted by age)
        # add result to dictionary
        age_dist[str(x) + '-' + str(x + bucket_size)], index = num_subj_in_bucket(df, len(df), index, x, bucket_size)
    
    return age_dist
```

```python
def df_from_dist(df, age_dist, age_dist_from, minimum = 0, maximum = 100, bucket_size = 5, min_percentage = 1):
    import math
    
    df_new = pd.DataFrame([])
    index = 0
    
    for x in np.arange(minimum, maximum, bucket_size):
        subj_bucket = []
        end_indx = age_dist[str(x) + '-' + str(x + bucket_size)]
    
        for i in range(end_indx):
            subj_bucket.append(df.iloc[index+i, 0])
        index = index + end_indx
    
        nndsp_bucket = age_dist_from[str(x) + '-' + str(x + bucket_size)] * min_percentage
        nndsp_bucket = math.ceil(nndsp_bucket)
    
        if not len(subj_bucket) == 0:
            if end_indx < nndsp_bucket:
                rand_subj_nki = np.random.choice(subj_bucket, size=end_indx, replace=False)
            else:
                rand_subj_nki = np.random.choice(subj_bucket, size=nndsp_bucket, replace=False)
        
            for sub in rand_subj_nki:
                #print(df_age_nki[df_age_nki.participant_id == sub])
                df_new = pd.concat([df_new, df[df.participant_id == sub]])
                
    return df_new
```

```python
def df_prop_dist(df, df_from, age_dist, age_dist_from, minimum = 0, maximum = 100, bucket_size = 5):
    min_percentage = 1.0
    percentages = []
    
    for x in np.arange(minimum, maximum, bucket_size):
        if age_dist_from[str(x) + '-' + str(x + bucket_size)] != 0:
            percent = age_dist[str(x) + '-' + str(x + bucket_size)] / age_dist_from[str(x) + '-' + str(x + bucket_size)]
            
            if percent > 0 and percent < min_percentage:
                percentages.append(percent)
                print(percent)
                min_percentage = percent
    
    if min_percentage < 0.1:
        min_percentage = 0.1
        
    print(min_percentage)
    
    return df_from_dist(df, age_dist, age_dist_from, minimum=minimum, maximum=maximum, 
                        bucket_size = bucket_size, min_percentage = min_percentage)
```

```python
def plot_dist(kwarg, ticks = False):
    for label, graph in kwarg.items():
        sns.distplot(graph, label = label) 
        
    if ticks:
        #plt.yticks(np.arange(0,0.05,0.0025)) # add y ticks
        plt.xticks(np.arange(0,100,5)) # add x ticks
    
    plt.legend(loc='best') # show legend
    plt.show()
```

```python
def plot_diff(a, b, bucket_size = 5, ticks = False):
    diff = []
    
    for key, num in a.items():
        diff.append(num - b[key])
        
    plt.scatter(np.arange(0,100,bucket_size), diff)
    
    if ticks:
        plt.xticks(np.arange(0,100,bucket_size))
    plt.show()
```

```python
def plot_count(kwarg, bucket_size = 5):
    for label, graph in kwarg.items():
        sns.distplot(graph, norm_hist = False, kde = False, bins = np.arange(0,100,bucket_size), label = label)
        
    plt.legend(loc='best')
    plt.xlabel('Age (Bin Size = ' + str(bucket_size) + ')')
    plt.ylabel('Count')
    plt.show()
```
